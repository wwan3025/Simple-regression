# start regression using statsmodels

## 
identify x and y variables, then run the regression, 
Tipically in regression, we also need a constant term which is called intercept for one on one linear regression 

```python
Xvar = df[['rooms']]
Yvar = df[['lprice']]

Xvar = sm.add_constant(Xvar)
```

```python
Yvar
```

```python
Xvar
```

# define the regression model

```python
# building OLS computing method within Statsmodels

reg = sm.OLS(Yvar, Xvar)

```

```python
# calcuate the regression model, in machine learning we call this training, in statistics we call this estimation, in applied mathematics, we call this optimisation 

# commond to train the model

result = reg.fit()
```
